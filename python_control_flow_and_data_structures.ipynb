# SECTION A — STRINGS

```python
# -----------------------------------
# Section A: String Manipulation (User Input)
# -----------------------------------
# This program demonstrates:
# - Converting text to title case
# - Accessing first and last characters
# - Reversing a string using a loop
# - Counting character occurrences
# - Removing spaces from a string
```

```python
# Asking user for their full name
name = input("Enter full name: ")

# convert name to title case and printing it
print(f"Name in title case: {name.title()}")

# printing the first and last character of the name.
print(f"First character: {name[0]}")
print(f"Last character: {name[-1]}")

# printing the name in reverse
# creating an empty string to put value in
reversedName = ""
# looping through all character in the name of the user
for i in name:
  # we are taking the first index of the user name and attaching that to the end of the new name.
  # So, in ABC, we are first writing A, and in this step we have reversedName empty, so it will stay A.
  # next, we are adding A to the end of the new i, which is B. So, we get    i + reversedName   as    "B" + "A" = BA (which is stored in reversedName)
  # next, we are adding B to the end of the new i, which is C. So, we get    i + reversedName   as    "C" + "BA" = CBA (now this is the new reversedName)
  reversedName = i + reversedName
print(f"Reversed Name: {reversedName}")

# Counting letter a
print(f"Total lower case 'a' in name: {name.lower().count("a")}")

# Removing all space from name
print(f"Without space: {name.replace(" ", "")}")
```

# SECTION B — NUMBERS, MATH & OPERATOR PRECEDENCE


```python
# -----------------------------------
# Exercise: Understanding Operator Precedence
# -----------------------------------
# This section evaluates different mathematical expressions to demonstrate:
# - Order of operations in Python
# - The effect of parentheses
# - Difference between division (/) and floor division (//)
# - Precedence of exponentiation over multiplication
```

```python
# First multiply 3 and 4, and then add the result to 4. Because multiplication has a higher precedence than addition
# 5 + 3 * 4     =>     5 + 12     =>     17
print(f"5 + 3 * 4: {5 + 3 * 4}")



# Parentheses have highest precedence, so we first do 5+3, and then we multiply the result with 4
# (5 + 3) * 4     =>     8 * 4     =>     32
print(f"(5 + 3) * 4: {(5 + 3) * 4}")



# For 100 / 6, standard division gives a float result
# 100 / 6     =>     16.666666666666668
# For 100 // 6, floor division gives the integer part only
# 100 // 6     =>     16
# python logic:
#     Evaluate
#     if A is false, return A
#     if A is true, return B
# thats why, in this case, even though both give slightly different answer (16.666666666666668 and 16), Python returns the value of the last one (which here is 16)
print(f"100 / 6 and 100 // 6: {100 / 6 and 100 // 6}")



# Exponentiation has higher precedence than multiplication
# So, we calculate like this:
# 2 ** 5 * 3     =>     32 * 3     =>     96
print(f"2 ** 5 * 3: {2 ** 5 * 3}")
```

# SECTION D — LOOPS

```python
# -----------------------------------
# Exercise: While Loop with break and continue
# -----------------------------------
# This program:
# - Iterates from 1 to 50
# - Skips numbers divisible by 4
# - Stops execution entirely when reaching 37
# - Demonstrates use of break and continue statements
```

```python
# starting from number 1
num = 1

# running a loop for 1 - 50
while num<=50:
  # running a if statement for 37, so that we can stop when it is encountered.
  if num == 37:
    break
  # skipping every 4th number, as in every number divisible by 4.
  if num % 4 == 0:
    num +=1
    continue
  # printing the number. Printing all number that were not skipped.
  print(num)
  # moving to the next number and continuing the loop
  num+=1
```

```python
# -----------------------------------
# Exercise: Multiplication Table Generator
# -----------------------------------
# This program:
# - Takes a user-entered number
# - Uses a for loop to generate its multiplication table (1–12)
```

```python
# get input from user
i = int(input("Enter value for multiplication table: "))

print(f"\nCalcalutaion for {i}'s table:\n")

# calculations (user-entered number   X   (1-12))
for x in range(1,13):

  # actual multiplcation for each line
  multiply = i*x

  # print each line in a multiplication table format
  print(f"{i} x {x} = {multiply}")
```

```python
# -----------------------------------
# Exercise: Nested Loop Pattern Printing
# -----------------------------------
# This section demonstrates nested loops by printing a numeric triangle pattern.
# 1
# 12
# 123
# 1234
# 12345
```

```python
# run a for-loop for number 1-100
for i in range(1,6):

  # creating an empty string. all the lines will be string instead of numbers.
  line = ""

  # loop for the triangle
  for j in range (1, i+1):

    # appending new string value to previous ones
    line = line + str(j)

  # printing the line, at the end of outermost loop
  print(line)
```


# SECTION E — LISTS & LIST COMPREHENSION

```python
# -----------------------------------
# Exercise: List Methods and Operations
# -----------------------------------
# This section demonstrates:
# - Sorting a list
# - Inserting elements at specific positions
# - Removing elements
# - Finding maximum and minimum values
# - Reversing a list
# - Counting occurrences of specific elements
```

```python
# List:
nums = [12, 4, 8, 1, 9, 3, 4, 4]

# sorting the list in ascending order
nums.sort()
print(f"Sorted list: {nums}")

# insert 100 in index 2
nums.insert(2,100)
print(f"After inserting 100 in index 2 we get: {nums}")

# removing value 1
nums.remove(1)
print(f"After removing 1 we get: {nums}")

# Finding and printing the max and min values
maxValue = max(nums)
minValue = min(nums)
print(f"Max Value: {maxValue}")
print(f"Min Value: {minValue}")


# Reversing list
nums.reverse()
print(f"Reversed list: {nums}")

# Counting how many times 4 appears
counting = nums.count(4)
print(f"Number of times 4 appeared: {counting}")
```

```python
# -----------------------------------
# Exercise: Conditional List Processing
# -----------------------------------
# This program:
# - Filters odd numbers
# - Identifies numbers greater than 20
# - Replaces numbers less than 10
```

```python
# created a list of 10 integers:
nums = [5, 12, 7, 25, 18, 3, 30, 9, 22, 1]

# printing the list
print(f"list: {nums}\n")

# printing only odd number
print("Odd Numbers:")
for i in nums:
  if i % 2 != 0:
    print(i)

# printing only numbers greater than 20
print("Numbers greater than 20: ")
for i in nums:
  if i > 20:
    print(i)

# replacing all numbers less than 10 with 0
print("After switching all numbers less than 10 with 0:")
for i in nums:
  if i >= 10:
    print(i)
  else:
    i = 0
    print(i)
```
